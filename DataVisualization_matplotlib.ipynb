````python
import matplotlib.pyplot as plt
import pandas as pd
player_data = pd.read_csv('nba_raw_players.csv')
player_id = pd.read_csv('nba_id_players.csv', header=None, names=['PLAYER_ID', 'NAME'])
position_data = pd.read_csv('nba_player_position.csv')

#%matplotlib inline #don't need to do plt.show()
````

## Line Plot

source: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html

syntax: `plt.plot(x_data, y_data, [color], [marker], [x2_data, y2_data,...])`

we can write `color = red, marker = 'o', linestyle = 'dashed'` as `fmt = '[marker][line][color]'=ro--`

````python
plt.plot(LeBron.SEASON_ID, LeBron.PTS, 'b')
plt.xlabel('SEASON')
plt.xticks(LeBron.SEASON_ID, rotation='vertical')
plt.title('LeBron James')
plt.tight_layout()
plt.savefig('graph/LeBron')
plt.show()
````

````python
plt.plot(LeBron.SEASON_ID, LeBron.PTS, 'b', label='LeBron')
plt.plot(Wade.SEASON_ID, Wade.PTS, 'r', label='Dwayne')
plt.xlabel('SEASON')
plt.xticks(Wade.SEASON_ID, rotation='vertical')
plt.title('LeBron James and Dwyane Wade')
plt.legend(loc = 'lower left')
plt.tight_layout()
plt.savefig('graph/two-line')
plt.show()
````

## using subplot()
syntax `subplot(nrows, ncols, nsubplot)`
subplot ordering:
1. Row-wise from top left 從左至右編號

|Left|Right|
|---|---|
|1|2|
|3|4|

2. Indexed from 1

`nrows`: number of rows

`ncols`: number of columns

`nsubplot`: ith subplot

````python
plt.subplot(2, 1 ,1)
plt.plot(LeBron.SEASON_ID, LeBron.PTS, 'b')
plt.xlabel('SEASON')
plt.xticks(LeBron.SEASON_ID, rotation='vertical')
plt.title('LeBron James')
plt.subplot(2, 1 ,2)
plt.plot(Wade.SEASON_ID, Wade.PTS, 'r')
plt.xlabel('SEASON')
plt.xticks(Wade.SEASON_ID, rotation='vertical')
plt.title('Dwyane Wade')
plt.tight_layout()
plt.savefig('separate.png')
plt.show()
````

## Add horizontal/vertical line and fill area

### Adding lines
syntax: `plt.axhline(y, c, ls, lw)` (horizontal line)

syntax: `plt.avhline(y, c, ls, lw)` (vertical line)

`y`: y position in data coordinates of the horizontal line.

`c`: color

`ls`: style of line

`lw`: width of line

`xmin`: Should be between 0 and 1, 0 being the far left of the plot, 1 the far right of the plot.

`xmax` : Should be between 0 and 1, 0 being the far left of the plot, 1 the far right of the plot.

### Filling area
syntax: `plot.fill(*args, data=None)`

### Filling area between lines

syntax: `plt.fill_between(x, y1, y2=0, where=None, interpolate=False, step=None)`

![alt](graph/10.png)

````python
plt.plot(LeBron.SEASON_ID, LeBron.PTS, 'b', label="LeBron James")
plt.legend()

plt.axhline(y=2000, c="r", ls="--", lw=2)
plt.xticks(LeBron.SEASON_ID, rotation='vertical')
plt.show()
````

````python
plt.plot(LeBron.SEASON_ID, LeBron.PTS, 'b', label="LeBron James")
plt.legend()

plt.axhline(y=2000, c="r", ls="--", lw=2)
#plt.fill_between(LeBron.SEASON_ID, LeBron.PTS, y2 = 2000, facecolor="lightgray")
plt.fill_between(LeBron.SEASON_ID, LeBron.PTS, y2 = 2000, where= LeBron.PTS >= 2000, facecolor="lightgray", interpolate=True)
plt.xticks(LeBron.SEASON_ID, rotation='vertical')
plt.show()
````

## Legends
to provide labels for overlaid points and curves
syntax: `plt.legend(loc, handles, label, )`

`loc`:
![alt](graph/05.png)

`bbox_to_anchor`: if we want to set the location of legend box, we can use this(https://www.longqi.cf/python/2014/11/12/LegendinMatplotlib/)

`handlers`:
>The original object which is used to generate an appropriate entry in the legend. ([Legend guide](https://matplotlib.org/3.1.1/tutorials/intermediate/legend_guide.html#sphx-glr-tutorials-intermediate-legend-guide-py))

Basically, the list we assigned decides whether an object is labeled and put in the legend.

`label`: if we want to change label

````python
plt.scatter(position_data[position_data['POSITION']=='Center'].HEIGHT, position_data[position_data['POSITION']=='Center'].WEIGHT, marker='x', label = 'Center', alpha = 0.2)
plt.scatter(position_data[position_data['POSITION']=='Forward'].HEIGHT, position_data[position_data['POSITION']=='Forward'].WEIGHT, marker='*', label = 'Forward', alpha = 0.3)
plt.scatter(position_data[position_data['POSITION']=='Guard'].HEIGHT, position_data[position_data['POSITION']=='Guard'].WEIGHT, label = 'Guard', alpha = 0.2)

plt.legend(loc = 'upper left')
plt.title('NBA position')
plt.xlabel('Height(")')
plt.ylabel('Weight(lb)')
plt.show()
````

## Histogram
syntax: `plot.hist(x, bins=None, range=None, density=None, cumulative=False, histtype='bar', align='mid', orientation='vertical', rwidth=None, color=None, label=None, stacked=False)`

`bin`: interval

`range` : is (x.min(), x.max()) by default

`density` : return percentage of every interval

`cumulative` : computed where each bin gives the counts in that bin plus all bins for smaller values

`histtype` : {‘bar’, ‘barstacked’, ‘step’, ‘stepfilled’}

`align` : {‘left’, ‘mid’, ‘right’}

`orientation` : {‘horizontal’, ‘vertical’}

`rwidth` : relative width of the bars as a fraction of the bin width

`color` : color

`label` : label

`stacked` : multiple data are stacked on top of each other

````python
plt.hist(position_data.HEIGHT, density=False, color = 'lightblue', cumulative = False, label = "Height")
plt.legend()
plt.xlabel('Height(")')
plt.savefig('graph/height')
plt.show()
````

````python
plt.hist([df1.HEIGHT, df2.HEIGHT, df3.HEIGHT], label = ['Center', 'Forward', 'Guard'], stacked=True)
plt.legend()
plt.show()
````

````python
plt.hist([Wade.PTS, LeBron.PTS], label = ['Dwayne Wade', 'LeBron James'], stacked=True)
plt.legend()
plt.show()
````

## Bar Chart
syntax: `plot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None)`

`x`: The x coordinates of the bars.

`height`: The height(s) of the bars.

`width`: The width(s) of the bars (default: 0.8).

`bottom`:The y coordinate(s) of the bars bases (default: 0).

`align` : {'center', 'edge'}, optional, default: 'center'. Alignment of the bars to the x coordinates

other parameters:

`color`, `ecolor`, `orientation`, `tick_label`, 

````python
plt.bar(position_data.POSITION.unique(),
        position_data.POSITION.value_counts(), 
        width=0.5, 
        bottom=None, 
        align='center', 
        color=['lightsteelblue', 
               'cornflowerblue', 
               'royalblue', 
               'midnightblue', 
               'navy', 
               'darkblue', 
               'mediumblue'],
#         orientation='horizontal'
       )
plt.xticks(rotation='vertical')
plt.axhline(y=100, c="r", ls="--", lw=2)

plt.show()
````

### plt.text()
syntax = `text(x_coordinate, y_coordinate, 'text_to_display', horizontalalignment='center', verticalalignment='center'`

````python
col_count = 3
bar_width = 0.2
index = np.arange(col_count)
A_scores = (553,536,548)
B_scores = (518,523,523)
C_scores = (613,570,588)
D_scores = (475,505,499)
````

````python
def createLabels(data):
    for item in data:
        height = item.get_height()
        plt.text(
            item.get_x()+item.get_width()/2., 
            height*1.05, 
            '%d' % int(height),
             ha = "center", # ha = horizontalalignment
             va = "bottom", # va = verticalalignment
        )
````

````python
A = plt.bar(index,
           A_scores, 
           bar_width,
           alpha=.4,
           label="K") 
B = plt.bar(index+0.2,
            B_scores,
            bar_width,
            alpha=.4,
            label="C") 
C = plt.bar(index+0.4,
            C_scores,
            bar_width,
            alpha=.4,
            label="N") # x,y ,width
D = plt.bar(index+0.6,
            D_scores,
            bar_width,
            alpha=.4,
            label="F") # x,y ,width

createLabels(A)
createLabels(B)
createLabels(C)
createLabels(D)

plt.ylabel("Mean score")
plt.xlabel("Subject")
plt.title("Test Scores by Contry")
plt.xticks(index + .3 / 2 + 0.15 ,("Math","Reading","Science"))
plt.legend() 
plt.grid(True) #add grid
plt.show()
````

## Box Plot
syntax: `plot.boxplot(x, notch=None, sym=None, vert=None, whis=None, positions=None, widths=None, usermedians=None, conf_intervals=None, labels=None, meanline=None, [other properties of box, fliers, ..etc])`

`x`: The input data.

`notch`: If True, will produce a notched box plot.

`sym`: The default symbol for flier points.

`vert`: If True (default), makes the boxes vertical. If False, everything is drawn horizontally.

`whis`: As a float, determines the reach of the whiskers to the beyond the first and third quartiles.

`usermedians`: An array or sequence whose first dimension (or length) is compatible with x. This overrides the medians computed by matplotlib for each element of usermedians that is not None.

`positions`: Sets the positions of the boxes.

`widths`: Sets the width of each box either with a scalar or a sequence.

`labels`: Labels for each dataset.

`meanline`:If True (and showmeans is True), will try to render the mean as a line spanning the full width of the box according to meanprops (see below). Not recommended if shownotches is also True. 

`boxprops, flierprops, medianprops, meanprops`: properties of elements in box plot

````python
plt.boxplot([position_data[position_data['POSITION']=='Center'].WEIGHT, 
             position_data[position_data['POSITION']=='Forward'].WEIGHT, 
             position_data[position_data['POSITION']=='Guard'].WEIGHT], 
            labels = ['Center', 'Forward', 'Guard'])
plt.ylabel('Weight(lb)')
plt.savefig('graph/box')
plt.show()
````

````python
plt.boxplot([position_data[position_data['POSITION']=='Center'].HEIGHT, 
             position_data[position_data['POSITION']=='Forward'].HEIGHT, 
             position_data[position_data['POSITION']=='Guard'].HEIGHT], 
            labels = ['Center', 'Forward', 'Guard'], notch = True, vert = True)
plt.show()
````

## Pie Chart
syntax: `plt.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False)`

`x`: The wedge sizes.

`explode`: If not None, is a len(x) array which specifies the fraction of the radius with which to offset each wedge.

`labels`: A sequence of strings providing the labels for each wedge

`colors`: Colors.

`autopct`: The label will be placed inside the wedge. If it is a format string, the label will be fmt%pct.

`pctdistance`: The ratio between the center of each pie slice and the start of the text generated by autopct.

`shadow`: Draw a shadow beneath the pie.

`labeldistance`: The radial distance at which the pie labels are drawn.

`radius`: The radius of the pie, if radius is None it will be set to 1.

`counterclock`: Specify fractions direction, clockwise or counterclockwise.

`wedgeprops`: Dict of arguments passed to the wedge objects making the pie.

`textprops`: Dict of arguments to pass to the text objects.

`center`: Center position of the chart. 

````python
plt.pie(position_data.POSITION.value_counts() , labels = position_data.POSITION.unique(),autopct='%1.1f%%')
plt.axis('equal')
plt.savefig('graph/pie')
plt.show()
````

````python
separated = (.1,0,0,0,0,0,0)
plt.pie(position_data.POSITION.value_counts() ,autopct='%1.1f%%',explode=separated, pctdistance = 0.6, textprops = {"fontsize" : 12})
plt.legend(labels = position_data.POSITION.unique(), loc='upper left')
plt.axis('equal')
plt.show()
````

````python
fig, ax = plt.subplots()

size = 0.3
vals = np.array([[position_data.POSITION.value_counts()['Guard'], position_data.POSITION.value_counts()['Guard-Forward']], 
                 [position_data.POSITION.value_counts()['Center'], position_data.POSITION.value_counts()['Center-Forward']], 
                 [position_data.POSITION.value_counts()['Forward'], position_data.POSITION.value_counts()['Forward-Guard']]])

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9, 10]))

ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.pie(vals.flatten(), radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='w'))
plt.legend(labels = position_data.POSITION.unique(), loc='upper right')
ax.set(aspect="equal", title='Pie plot with position')
plt.show()
````
